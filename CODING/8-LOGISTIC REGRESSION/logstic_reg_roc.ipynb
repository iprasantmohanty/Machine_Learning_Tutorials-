```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_curve,roc_auc_score
%matplotlib inline
```

## create data set

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_classes=2,random_state=1)
```

```python
pd.DataFrame(X)
```

```python
pd.DataFrame(y)
```

## train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

creating dummy model with default 0 as output

```python
dummy_model_prob=[0 for _ in range(len(y_test))]
dummy_model_prob
```

## model training

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

## prediction based on probability

```python
model_prob=model.predict_proba(X_test)
```

```python
model_prob
```

```python
model_prob=model_prob[:,1]
model_prob
```

## calculate the score

```python
dummy_model_auc=roc_auc_score(y_test,np.array(dummy_model_prob))
model_auc=roc_auc_score(y_test,np.array(model_prob))
print(dummy_model_auc)
print(model_auc)
```

## ROC Curve

```python
dummy_fpr,dummy_tpr,_=roc_curve(y_test,np.array(dummy_model_prob))
model_fpr,model_tpr,thresholds=roc_curve(y_test,model_prob)
```

```python
thresholds
```

```python
plt.plot(dummy_fpr,dummy_tpr,linestyle='--',label='Dummy Model')
plt.plot(model_fpr,model_tpr,marker='.',label='Logistic')
plt.xlabel('Flase Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()

```

```python
fig=plt.figure(figsize=(20,50))
plt.plot(dummy_fpr,dummy_tpr,linestyle='--',label='Dummy Model')
plt.plot(model_fpr,model_tpr,marker='.',label='Logistic')
ax=fig.add_subplot(111)
for xyz in zip(model_fpr,model_tpr,thresholds):
    ax.annotate('%s' % np.round(xyz[2],2),xy=(xyz[0],xyz[1]))
plt.xlabel('Flase Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

## Hyperparameter Tuning

```python
model=LogisticRegression()
penalty=['l1', 'l2', 'elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

### grid search CV

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
cv=StratifiedKFold()
```

```python
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

### Randomized Search CV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model=LogisticRegression()
```

```python
randomcv=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy')
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_params_
```

```python
randomcv.best_score_
```

```python
y_pred=randomcv.predict(X_test)
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
