```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
df=pd.DataFrame(dataset.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
X=df
y=dataset['target']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
import seaborn as sns
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),annot=True)
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_reg=DecisionTreeRegressor()
```

```python
dt_reg.fit(X_train,y_train)
```

```python
y_pred=dt_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

## Hyperparameter Tuning

```python
params={
    'criterion' :["squared_error","friedman_mse","absolute_error"],
    'splitter':["best", "random"],
    'max_depth': [1,2,3,4,5,10,15,20,25],
    'max_features' : ['auto',"sqrt", "log2"]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
reg=DecisionTreeRegressor()
```

```python
grid=GridSearchCV(reg,param_grid=params,cv=5,scoring='neg_mean_squared_error')
```

```python
## to ignore/remove warnings
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
grid.best_params_
```

```python
selected_model=DecisionTreeRegressor(criterion='friedman_mse',max_depth=3,max_features='log2',splitter='best')
```

```python
selected_model.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selected_model,filled=True)
```
